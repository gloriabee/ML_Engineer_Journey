### Linear Regression Equation

**In algebraic terms**

```
y= mx+b
```


*   y is miles per gallon-the value we want to predict.
*   m is the slope of the line.
*   x is pounds—our input value.
*   b is the y-intercept.


**In ML terms**

```
y'= b + w1x1
```


*   y' is the predicted label- the output.
*   b is the bias of model.(parameter of model and is calculated during training)
*   w1 is weight of feature.(parameter of model and is calculated during training)
*   x1 is a feature- the input.
